### 0px로 된 데이터 카운트

```python
print(sum(wines.bodyPercent == "0px;")) # 0개
print(sum(wines.dryOrSweet == "0px;")) # 64개
print(sum(wines.Acidic == "0px;")) # 0개
```

```python
wines.dryOrSweet[wines.dryOrSweet == "0px;"] = "0%;"
```

```python
wines.dryOrSweet[wines.dryOrSweet == "0%;"]
```

```python
newBodyPercent = []
newDryOrSweet = []
newAcidic = []


for i in wines.bodyPercent:
    newBodyPercent.append(i[:-2])
wines['NewBodyPercent'] = newBodyPercent


for i in wines.dryOrSweet:
    newDryOrSweet.append(i[:-2])
wines['NewDryOrSweet'] = newDryOrSweet


for i in wines.Acidic:
    newAcidic.append(i[:-2])

wines['NewAcidic'] = newAcidic

wines
```

```python
df = pd.read_csv("./Data/French_red_white_wines.csv")
```

```python
len(df.about_wine)
```

```python
grape_list = []
alcohol_list = []
allergy_list = []


for i in df.about_wine:
    # print(i.split("[")[3].replace("'", "").replace("]", "").split(",")[1:].replace())
    try:
        grape_list.append(i.split("[")[3].replace("'", "").replace("]", "").split(",")[1:])
    except:
        grape_list.append("")

for i in df.about_wine:
    try:
        alcohol_list.append(i.split("[")[6].split(',')[1].replace("'", "").replace("]", ""))
    except:
        alcohol_list.append("")

for i in df.about_wine:
    try:
        allergy_list.append(i.split("[")[7].split(",")[1:])
    except:
        allergy_list.append("")
    

# df



df['Grape_Type'] = grape_list
df['Alcohol'] = alcohol_list
df['Allergy'] = allergy_list
df.to_csv("./Data/French_red_white_wines.csv", index=None)
```
