```python
def compareFeatureShift(df):
    return df[(df[['bodyPercent', 'Tanning', 'Sugar', 'Acidic']] == df[['bodyPercent', 'Tanning', 'Sugar', 'Acidic']].shift(-1)).all(axis=1)]
```

```python
sameindex = list(compareFeatureShift(rwbg).index.values)
```

```python
rwbg = rwbg.drop(index=sameindex).reset_index(drop=True)
```

```python
rwbg
```

```python
def inner_join(leftdf, rightdf):
    right_copy = rightdf.drop_duplicates()
    right_copy.drop(columns=['Allergy', 'bodyPercent', 'Tanning', 'Acidic', 'NewBodyPercent', 'NewTanning', 'NewAcidic', 'year', 'price'], inplace=True)
    right_copy.reset_index(drop=True, inplace=True)
    left_copy = leftdf.drop_duplicates()
    inner_join_df = pd.merge(left_copy, right_copy, on='name', how='left')
    inner_join_df.dropna(axis=0, inplace=True)
    inner_join_df.reset_index(drop=True, inplace=True)

    return inner_join_df
```

```python
rwbg = inner_join(rwbg, rwall)
```

```python
# Tasting 0px; 정리 함수
def getOnlyNumbersTasting(df):
    # df : 변경하고자 하는 DataFrame
    print(f"body Percent의 0px; count : {sum(df.bodyPercent == '0px;')}")
    print(f"Tanning의 0px; count : {sum(df.Tanning == '0px;')}")
    print(f"Sugar의 0px; count : {sum(df.Sugar == '0px;')}")
    print(f"Acidic의 0px; count : {sum(df.Acidic == '0px;')}")
    df.bodyPercent[df.bodyPercent == "0px;"] = "0%;"
    df.Tanning[df.Tanning == "0px;"] = "0%;"
    df.Sugar[df.Sugar == "0px;"] = "0%;"
    df.Acidic[df.Acidic == "0px;"] = "0%;"
    df['NewBodyPercent'] = [i[:-2] for i in df.bodyPercent]
    df['NewTanning'] = [i[:-2] for i in df.Tanning if pd.notna(i)]
    df['NewDryOrSweet'] = [i[:-2] for i in df.Sugar]
    df['NewAcidic'] = [i[:-2] for i in df.Acidic]

    return df
```

```python
getOnlyNumbersTasting(wines)
```
